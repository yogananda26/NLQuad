```python
from datasets import load_dataset

raw = load_dataset("json", data_files="/content/NLQuAD_train.json")
```

```python
def flatten(example):
    rows = []

    title = example["title"]
    date = example["date"]

    for para in example["paragraphs"]:
        context = para["context"]

        for qa in para["qas"]:
            question = qa["question"]
            qa_id = qa["id"]

            for ans in qa["answers"]:
                rows.append({
                    "title": title,
                    "date": date,
                    "context": context,
                    "question": question,
                    "answer_text": ans["text"],
                    "answer_start": ans["answer_start"],
                    "answer_end": ans["answer_end"],
                    "qa_id": qa_id
                })

    return rows
```

```python
print(flatten(raw["train"][0]))
```

```python
dataset = raw["train"].map(
    flatten,
    batched=False,
    remove_columns=raw["train"].column_names
)
```
